# Collating all the data for Penoyre+ 18

```python
%load_ext autoreload
%autoreload 2
import numpy as np
import astropy
import SEDBuilder.spectra
import astroquery

SEDBuilderDir=SEDBuilder.__file__[:SEDBuilder.__file__.rfind('/')] # (just in case you're running this in another directory)
dataDir=SEDBuilderDir+'/spectraData'
```

## Furlan+ 2011 
(https://ui.adsabs.harvard.edu/?#abs/2011ApJS..195....3F/abstract)

#### An unbiased selection of 161 YSOs in Taurus - also contains data about binaries - no SED data points

```python
source='Furlan+ 2011'

furlanData=astropy.io.ascii.read('http://iopscience.iop.org/0067-0049/195/1/3/suppdata/apjs393993t1_ascii.txt',header_start=2,data_start=4,data_end=-10,delimiter='\t')
multipleData=astropy.io.ascii.read('http://iopscience.iop.org/0067-0049/195/1/3/suppdata/apjs393993t3_ascii.txt',header_start=2,data_start=5,data_end=-17,delimiter='\t')

# finding the names of binaries in the sample
mNames=multipleData['Name']
for i in range(mNames.size):
    mName=mNames[i]
    # some stars have a chevron in their name
    chevPos=mName.find('^')
    if chevPos!=-1:
        mNames[i]=mName[:chevPos]

nStars=len(furlanData)
for i in range(nStars):
    print('star ',i,' of ',nStars)
    
    ra=furlanData['R.A. (J2000)'][i]
    dec=furlanData['Decl. (J2000)'][i]
    name=furlanData['Name'][i]
    spType=str(furlanData['Spectral Type'][i]) # astropy doesn't like writing numpy._str to file
    
    starInfo=SEDBuilder.spectra.getInfo(ra=ra,dec=dec)
    twoMassId=starInfo['2MASSID']
    SEDBuilder.spectra.getSpectra(name=starInfo['SimbadName'],saveDir=dataDir,overwrite=1)
    
    SEDBuilder.spectra.addRegionToMetadata(twoMassId,dataDir,'Taurus',source)
    SEDBuilder.spectra.addToMetadata(twoMassId,dataDir,'StellarType',spType,source)
    
    if name in mNames:
        SEDBuilder.spectra.addToMetadata(twoMassId,dataDir,'BinaryFlag',1,source)
    else:
        SEDBuilder.spectra.addToMetadata(twoMassId,dataDir,'BinaryFlag',0,source)
    
    
```

# Manoj+ 2011

(https://ui.adsabs.harvard.edu/?#abs/2011ApJS..193...11M/abstract)

#### An unbiased selection of 84 YSOs in Cha I (though will store their region as Chamaeleon - note the spelling)

```python
source='Manoj+ 2011'

manojData=astropy.io.ascii.read('http://iopscience.iop.org/0067-0049/193/1/11/suppdata/apjs382407t1_ascii.txt',header_start=2,data_start=5,data_end=-1,delimiter='\t')

nStars=len(manojData)

for i in range(nStars):
    print('star ',i,' of ',nStars)
    
    manojName='2MASS '+manojData['2MASS ID'][i] # anyone who lists the 2MASS Id of all stars in their survey is a hero
    starInfo=SEDBuilder.spectra.getInfo(name=manojName)
    twoMassId=starInfo['2MASSID']
    
    SEDBuilder.spectra.getSpectra(name=starInfo['SimbadName'],saveDir=dataDir,overwrite=1)
    
    SEDBuilder.spectra.addRegionToMetadata(twoMassId,dataDir,'Chamaeleon',source)
```

# McClure+ 2010
(https://ui.adsabs.harvard.edu/?#abs/2010ApJS..188...75M/abstract)

#### 136 YSOs in Ophiuchus (this table proved difficult to read so thanks to Emily Sandford for writing a modification to astropy that could manage it)

```python
# These 3 functions are just for reading this particular data set
def read_oph_table(table, delimiter='\t'):
    reader = astropy.io.ascii.Basic()
    reader.header.splitter.delimiter = delimiter
    reader.data.splitter.delimiter = delimiter
    reader.header.splitter.process_line = process_header_line
    reader.data.splitter.process_line = process_data_line
    reader.header.start_line = 2
    reader.data.start_line = 3
    
    return reader.read(table)
def process_data_line(line, desired_ncol=9):
        """
        If there are already 9 columns in this line of data, proceed as usual.
        
        If there are only 8, add another column that just says "None."
        """
        if (len(line.strip().split("\t"))==desired_ncol):
            return line.strip()
        else:
            return "\t".join(line.strip().split("\t")) + "\tNone"
def process_header_line(line, desired_ncol=9):
    """
    Get rid of the last, empty column header.
    """
    if (len(line.strip().split("\t"))==desired_ncol):
        return line.strip()
    else:
        return "\t".join(line.strip().split("\t")[:-1])

source='McClure+ 2010'
mcclureData = read_oph_table('http://iopscience.iop.org/0067-0049/188/1/75/suppdata/apjs330182t1_ascii.txt')

realStars=np.flatnonzero(mcclureData['Comments']=='None') # At time of writing this comment, I can't remember why this line is here. A mystery worth investigating?
nStars=len(realStars)

for j in range(nStars):
    print(j,' of ',nStars,' stars')
    i=realStars[j]
    if (mcclureData['Name'][i]=='VLA-1623'):
        continue # not a star apparently
    ra=mcclureData['R.A. (J2000)'][i]
    dec=mcclureData['Decl. (J2000)'][i]
    
    starInfo=SEDBuilder.spectra.getInfo(ra=ra,dec=dec)
    twoMassId=starInfo['2MASSID']
    
    SEDBuilder.spectra.getSpectra(name=starInfo['SimbadName'],saveDir=dataDir,overwrite=1)
    
    SEDBuilder.spectra.addRegionToMetadata(twoMassId,dataDir,'Ophiuchus',source)
    
```

# Andrews+ 2013
(https://ui.adsabs.harvard.edu/#abs/2013ApJ...771..129A/abstract)

#### 210 YSOs in Taurus - SED data is machine readable (thus found by getSpectra() already)
#### Also contains Teff, Av, Mass and Age

Note - This dataset is really patchy in it's book-keeping. Each table has a different length, it's full of binaries though they're not handled very consistently, and many of the names used are not recognisable. I've done my best to smooth over all those issues without throwing away data, but I bet some oddities have slipped through the cracks.

```python
source='Andrews+ 2013'

tab2=astroquery.vizier.Vizier.get_catalogs('J/ApJ/771/129/table2')[0]
tab3=astroquery.vizier.Vizier.get_catalogs('J/ApJ/771/129/table3')[0]
tab4=astroquery.vizier.Vizier.get_catalogs('J/ApJ/771/129/table4')[0]

names=tab2['Name']
spTypes=tab2['SpT']

nStars=names.size
for i in range(nStars):
    name=names[i]
    binaryFlag=0
    print('_________star ',i,' of ',nStars)
    print('name: ',name)
    if (name[-2:]==' A') | (name[-3:]==' AB') | (name[-3:]=='Aab'): #this is a binary but lets just take info about the primary
        binaryFlag=1
        name=name[:name.rfind(' ')]
        print('__________',name)
    elif (name[-2:]==' B') | (name[-2:]=='Ba')| (name[-2:]=='Bb') | (name[-2:]==' C'):
        continue # skipping secondarys
 
    if name[:2]!='FV': # Don't even ask...
        name=name.replace('/',' ')
    if name[:2]=='J0':
        name='2MASS '+name
    if name[:4]=='CIDA':
        name='[BCG93]'+name[4:]
    if name[:4]=='KPNO':
        name='[BLH2002] KPNO-Tau'+name[-2:]
    if name[:4]=='V410':
        name='[SS94]'+name
    if name=='UX Tau':
        name=name+' A' # The main system doesn't have a 2MASS entry
    if name[:6]=='UZ Tau':
        name=name[:6]+' A' # The main system doesn't have a 2MASS entry
    if name=='Haro 6-37':
        name=name+' A' # The main system doesn't have a 2MASS entry
    if (name=='T Tau N') | (name=='FU Tau') | (name=='V710 Tau') | (name=='CFHT 4') | (name=='St 34'):
        continue # these star don't appear to have a 2MASS name so skipping
        
    print('now name is: ',name)
    spType=str(spTypes[i])
    print('spType: ',spType)
    
    table=SEDBuilder.spectra.getSpectra(name=name,saveDir=dataDir,overwrite=0)
    twoMassId=table.meta['2MASSID']
    
    SEDBuilder.spectra.addRegionToMetadata(twoMassId,dataDir,'Taurus',source)
    SEDBuilder.spectra.addToMetadata(twoMassId,dataDir,'StellarType',spType,source)
    
    # Check if theres data about this object in tbales 3 and 4 (which seem to have a different list of stars...)
    tab3entry=-1
    if name in tab3['Name']:
        tab3entry=np.flatnonzero(tab3['Name']==name)[0]
    elif names[i] in tab3['Name']:
        tab3entry=np.flatnonzero(tab3['Name']==names[i])[0]
    tab4entry=-1
    if name in tab4['Name']:
        tab4entry=np.flatnonzero(tab4['Name']==name)[0]
    elif names[i] in tab4['Name']:
        tab4entry=np.flatnonzero(tab4['Name']==names[i])[0]
        
    # Note Andrews+ use 3 different stellar models to produce 3 different (though relatively consistent) fits to mass and age
    # Here we use their 3rd quoted values, the fit to the most recent model (2000A&A...358..593S) - generally seems to give median values
    if tab3entry!=-1:
        mass=10**tab3['logM_3'][tab3entry]
        age=10**tab3['logt_3'][tab3entry]
        print('mass: ',mass)
        print('age: ',age)
        SEDBuilder.spectra.addToMetadata(twoMassId,dataDir,'Mass',mass,source)
        SEDBuilder.spectra.addToMetadata(twoMassId,dataDir,'Age',age,source)
    if tab4entry!=-1:
        teff=10**tab4['logT_'][tab4entry]
        lum=10**tab4['logL_'][tab4entry]
        print('teff: ',teff)
        print('lum: ',lum)
        SEDBuilder.spectra.addToMetadata(twoMassId,dataDir,'Teff',teff,source)
        SEDBuilder.spectra.addToMetadata(twoMassId,dataDir,'Luminosity',lum,source)
    
```

# Ribas+ 2017
(https://ui.adsabs.harvard.edu/#abs/2017ApJ...849...63R/abstract)

#### Herschel data for 283 YSOs in Taurus, Ophiuchus and Chamealeon
#### Includes stellar types, temperatures, luminosities and Avs

```python
source='Ribas+ 2017'

table=astropy.io.ascii.read('http://iopscience.iop.org/0004-637X/849/1/63/suppdata/apjaa8e99t15_mrt.txt')

#for key in table.keys():
#    print(key)
    
raHs=table['RAh']
raMs=table['RAm']
raSs=table['RAs']
decSigns=table['DE-']
decDs=table['DEd']
decMs=table['DEm']
decSs=table['DEs']

spTypes=table['SpType']
temps=table['Teff']
lums=table['L*']
avs=table['AV']
regions=table['Region']

nStars=raHs.size
for i in range(nStars):
    print(i,' of ',nStars)
    ra=str(raHs[i])+':'+str(raMs[i])+':'+str(raSs[i])
    dec=str(decSigns[i])+str(decDs[i])+':'+str(decMs[i])+':'+str(decSs[i])
    
    spectra=SEDBuilder.spectra.getSpectra(ra=ra,dec=dec,saveDir=dataDir,overwrite=0)
    twoMassId=spectra.meta['2MASSID']
    
    SEDBuilder.spectra.addRegionToMetadata(twoMassId,dataDir,regions[i],source)
    SEDBuilder.spectra.addToMetadata(twoMassId,dataDir,'StellarType',str(spTypes[i]),source)
    SEDBuilder.spectra.addToMetadata(twoMassId,dataDir,'Teff',temps[i],source)
    SEDBuilder.spectra.addToMetadata(twoMassId,dataDir,'Luminosity',lums[i],source)
    SEDBuilder.spectra.addToMetadata(twoMassId,dataDir,'Av',avs[i],source)
    
    ls=np.array([7e-5,1e-4,1.6e-4,2.5e-4,3.5e-4,5e-4])
    fs=np.array([table['F70'][i],table['F100'][i],table['F160'][i],table['F250'][i],table['F350'][i],table['F500'][i]])
    es=np.array([table['e_F70'][i],table['e_F100'][i],table['e_F160'][i],table['e_F250'][i],table['e_F350'][i],table['e_F500'][i]])
    fs=fs*3e8*1e-29/(ls**2)
    es=es*3e8*1e-29/(ls**2)
    
    nonZeros=np.flatnonzero(fs>0 & np.isfinite(fs))
    SEDBuilder.spectra.addToSpectra(twoMassId,dataDir,ls[nonZeros],fs[nonZeros],es[nonZeros],source,'Herschel')

    
```

# Bulger+ 2014
(https://ui.adsabs.harvard.edu/#abs/2014A&A...570A..29B/abstract)

#### 134 low mass YSOs in Taurus - SED data already in Vizier, but also has stellar types, radii, temperatures and Avs

```python
source = 'Bulger+ 2014'

table=astroquery.vizier.Vizier.get_catalogs('J/A+A/570/A29/table4')[0]
tmNames=table['_2MASS']
avs=table['Av1']
rs=table['Rs1']
temps=table['Teff']
spTypes=table['SpType']

nStars=len(table)
skipIndices=[55,151,154,155,156,160]
for i in range(nStars):
    print('star ',i,' of ',nStars)
    if i in skipIndices:
        continue # data with no 2MASS name
    name='2MASS'+tmNames[i]
    print('name: ',name)
    
    spectra=SEDBuilder.spectra.getSpectra(name=name,saveDir=dataDir,overwrite=0)
    twoMassId=spectra.meta['2MASSID']
    
    SEDBuilder.spectra.addRegionToMetadata(twoMassId,dataDir,'Taurus',source)
    if spTypes[i]!='':
        SEDBuilder.spectra.addToMetadata(twoMassId,dataDir,'StellarType',str(spTypes[i]),source)
    if np.ma.is_masked(temps[i])==False:
        SEDBuilder.spectra.addToMetadata(twoMassId,dataDir,'Teff',temps[i],source)
    if np.ma.is_masked(rs[i])==False:
        SEDBuilder.spectra.addToMetadata(twoMassId,dataDir,'Radius',rs[i],source)
    if np.ma.is_masked(avs[i])==False:
        SEDBuilder.spectra.addToMetadata(twoMassId,dataDir,'Av',avs[i],source)
    
    
```

# Herczeg+ 2014
(https://ui.adsabs.harvard.edu/?#abs/2014ApJ...786...97H)

#### Supplementary information for 399 stars in various regions

```python
source='Herczeg+ 2014'

table=astroquery.vizier.Vizier.get_catalogs('J/ApJ/786/97/table14')[0]
tmNames=table['OName']
spTypes=table['SpType']
avs=table['AV']
masses=table['Mass']
ages=np.power(10,table['logA']-6) # in Myr?
rads=table['Rad']
lums=np.power(10,table['logL'])
nStars=len(table)

skipStars=[58] #not actual usable designations
for i in range(224,nStars):
    print(i,' of ',nStars)
    name=tmNames[i]
    if (name[0]!='2') | (i in skipStars):
        continue
    
    spectra=SEDBuilder.spectra.getSpectra(name=name,saveDir=dataDir,overwrite=0)
    twoMassId=spectra.meta['2MASSID']
    
    SEDBuilder.spectra.addToMetadata(twoMassId,dataDir,'StellarType',str(spTypes[i]),source)
    SEDBuilder.spectra.addToMetadata(twoMassId,dataDir,'Luminosity',lums[i],source)
    SEDBuilder.spectra.addToMetadata(twoMassId,dataDir,'Radius',rads[i],source)
    SEDBuilder.spectra.addToMetadata(twoMassId,dataDir,'Av',avs[i],source)
    
    if masses[i]>0:
        SEDBuilder.spectra.addToMetadata(twoMassId,dataDir,'Mass',masses[i],source)
        SEDBuilder.spectra.addToMetadata(twoMassId,dataDir,'Age',ages[i],source)
```
